```python
import numpy as np

# Replace ? with NaN across the entire dataframe
df = df.replace('?', np.nan)

# Drop columns with too much missing data
df = df.drop(columns=['weight', 'payer_code', 'medical_specialty'])

# Confirm the drops worked
print(df.shape)
```

```python
# Drop high missing value columns
df = df.drop(columns=['max_glu_serum', 'A1Cresult'])

# Confirm
print(df.shape)
```

```python
# Replace NaN in race with 'Unknown'
df['race'] = df['race'].fillna('Unknown')

# Confirm no more nulls in race
print(df['race'].isnull().sum())
```

```python
age_mapping = {
    '[0-10)': 5,
    '[10-20)': 15,
    '[20-30)': 25,
    '[30-40)': 35,
    '[40-50)': 45,
    '[50-60)': 55,
    '[60-70)': 65,
    '[70-80)': 75,
    '[80-90)': 85,
    '[90-100)': 95
}

df['age'] = df['age'].map(age_mapping)
print(df['age'].dtype)
print(df['age'].value_counts())
```

```python
readmitted_mapping = {
    '<30': 1,
    '>30': 0,
    'NO': 0
}

df['readmitted'] = df['readmitted'].map(readmitted_mapping)

# Confirm
print(df['readmitted'].value_counts())
```
